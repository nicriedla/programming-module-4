```python
import matplotlib.pyplot as plt
# Seus dados
dados = [
    (0, 5.00, 0.00),
    (401, 4.80, 0.20),
    (803, 4.61, 0.39),
    (1205, 4.43, 0.57),
    (1607, 4.26, 0.74),
    (2010, 4.09, 0.91),
    (2412, 3.93, 1.07),
    (2813, 3.77, 1.23),
    (3216, 3.63, 1.37),
]

# Separando os dados
x = [item[0] for item in dados]
y1 = [item[1] for item in dados]
y2 = [item[2] for item in dados]

# Criando o grfico
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Tenso no C (Vc)')
# Configurando rtulos e ttulo
plt.xlabel('Tempo (ms)')
plt.ylabel('Tenso (V)')
plt.title('Carga no Capacitor (C)')
plt.legend()
# Exibindo o grfico
plt.grid(True)
plt.show()

# Criando o grfico
plt.figure(figsize=(10, 6))
plt.plot(x, y2, label='Tenso no R (Vr)', color='red')
# Configurando rtulos e ttulo
plt.xlabel('Tempo (ms)')
plt.ylabel('Tenso (V)')
plt.title('Descarga no Resistor (R)')
plt.legend()
# Exibindo o grfico
plt.grid(True)
plt.show()

# Criando o grfico
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Tenso no C (Vc)')
plt.plot(x, y2,label='Tenso no R (Vr)', color='red' )
# Configurando rtulos e ttulo
plt.xlabel('Tempo (ms)')
plt.ylabel('Tenso (V)')
plt.title('Comparao: Carga no C e Descarga no R')
plt.legend()
# Exibindo o grfico
plt.grid(True)
plt.show()
```
